```ocaml
let pred i = (i - 1)/2;;
let swap heap i j =
    let tmp = heap.a.(i) in
    heap.a.(i) <- heap.a.(j);
    heap.a.(j) <- tmp;;
```

```ocaml
let rec up heap i =
    let p = pred i in
    if i <> 0 && heap.a.(p) < heap.a.(i) then (
        swap heap i p;
        up heap p
    )
```

```ocaml
let rec down heap i =
    let get j = (if j < heap.n then heap.a.(j) else min_int), j in
    let m, j = max (get (2*i + 1)) (get (2*i + 2)) in
    if heap.a.(i) < m then (
        swap heap i j;
        down heap j
    )
```

```ocaml
let add heap e =
    heap.a.(heap.n) <- e;
    up heap heap.n;
    heap.n <- heap.n + 1;;
```

```ocaml
let array_to_heap a =
    let heap = { a=a; n=0 } in
    Array.iter (add heap) a;
    heap;;
    
let h = array_to_heap [|4; 1; 2; 8; 6; 3|];;
draw_heap ~file_out:"heap_ex" h
```

```ocaml
let take_max heap =
    swap heap 0 (heap.n - 1);
    heap.n <- heap.n - 1;
    down heap 0;
    heap.a.(heap.n);;
    
take_max h;;
```
